## 3. ***Data Wrangling***

### Data Wrangling Code

Firstly, we need to convert the data types with their respective actual type.

Reviews, Size, Installs, & Price should have an integer or float datatype, But here we can see of object type, So let's convert them to their respective type.

````python
#Converting the Reviews column to integer.
Play_Store_Data['Reviews'] = Play_Store_Data.Reviews.astype(int)
````

````python
#remove (+) and (,) from install
Play_Store_Data['Installs'] = Play_Store_Data.Installs.apply(lambda x: x.strip('+')) #strip():trim char from both ends of the string
Play_Store_Data['Installs'] = Play_Store_Data.Installs.apply(lambda x: x.replace(',', ''))
````

````python
#for convert the numeric type
Play_Store_Data['Installs'] = pd.to_numeric(Play_Store_Data['Installs']) # Converting the string to Numeric type
````

In size we know that 1 MB = 1024 KB ,to convert KB to MB, we must divide all the values which are in KB by 1024.

````python
#convert KB into the MB.
Play_Store_Data['Size'] = Play_Store_Data.Size.apply(lambda x: x.rstrip('M'))#rstrip():trim char from right side of a string.
Play_Store_Data['Size'] = Play_Store_Data.Size.apply(lambda x: float(x.rstrip('k').rstrip('K'))/1024 if 'k' in x or 'K' in x else x)
````

````python
Play_Store_Data.loc[10468:10477, :]
````

````python
Play_Store_Data['Size'] = Play_Store_Data.Size.replace('Varies with device', np.NaN)
````

````python
Play_Store_Data['Size'] = pd.to_numeric(Play_Store_Data['Size'])
````

````python
#remove $ sign from the price
Play_Store_Data['Price'] = Play_Store_Data.Price.apply(lambda x: x.strip('$'))
````

````python
Play_Store_Data['Price'] = pd.to_numeric(Play_Store_Data['Price'])
````
